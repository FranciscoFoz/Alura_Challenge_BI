## Funções

```python
def tabela_frequencia(variavel):
  '''
  Esta função irá gerar uma tabela de frequência com percentuais de acordo 
  com a variável passada. 

  variavel = variável categórica escolhida de dentro do conjunto de dados 
  emprestimos_completo
  '''

  dataframe = pd.DataFrame(emprestimos_completo[variavel].value_counts())                      
  dataframe.columns = ['quantidade']
  dataframe['percentual'] = round((dataframe.quantidade / dataframe.quantidade.sum())*100,4)

  return dataframe
```

```python
def tabela_frequencia_por_ano(variavel):
  '''
  Esta função irá gerar uma tabela de frequência com percentuais de acordo 
  com a variável passada. 

  variavel = variável categórica escolhida de dentro do conjunto de dados 
  emprestimos_completo
  '''

  dataframe = emprestimos_completo.loc[:,['data_emprestimo', variavel]]
  dataframe['data_emprestimo'] = dataframe['data_emprestimo'].dt.year
  dataframe = pd.DataFrame(dataframe.value_counts().groupby(by=['data_emprestimo',variavel]).sum(),columns=['quantidade'])
  dataframe.index.names = ['ano',variavel]
  dataframe = dataframe.reset_index().pivot(index=variavel,columns='ano')

  return dataframe
  
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def gera_box_plot(dataset,x,y,titulo):
  '''
  Esta função irá gerar um gráfico de boxplot.

  Dataset = conjunto de dados do gráfico
  x = valor do eixo x do gráfico
  y = valor do eixo y do gráfico
  título = título do gráfico
  '''

  sns.set_theme(style="darkgrid", palette='coolwarm',font_scale=1.3)                    #Definir a cor do fundo, paleta de cores e fonte dos eixos
  plt.figure(figsize=(16,10))                                                           #Dimensionar o tamanho

  ax = sns.boxplot(y= y, x= x, data= dataset)                                           #Plotar o gráfico

  plt.ylim(0,max(dataset[y])*1.1)                                                       #Definir o limite do eixo y
  plt.xlabel(None)                                                                      #Excluir título do eixo x
  plt.ylabel(None)                                                                      #Excluir título do eixo x

  plt.title(titulo,loc='left',size=20)                                                  #Gerar o título
  

  
```

```python
def gera_tabela_emprestimos_por_CDU(CDU):
  '''
  Esta função irá gerar a tabela da quantidade de empréstimos por alunos de 
  graduação, nos anos de 2017,2018 e 2019 (definida pela variável 
  zila_mamede_emprestimo_aluno_graduacao) filtrada pela CDU escolhida.

  CDU = Classificação geral da CDU definida na tabela
  '''

  
  zila_mamede_emprestimo_aluno_graduacao_CDU =  zila_mamede_emprestimo_aluno_graduacao[zila_mamede_emprestimo_aluno_graduacao['CDU_geral']==CDU]
  zila_mamede_emprestimo_aluno_graduacao_CDU = zila_mamede_emprestimo_aluno_graduacao_CDU.loc[:,['data_emprestimo']]
  zila_mamede_emprestimo_aluno_graduacao_CDU['ano'] = zila_mamede_emprestimo_aluno_graduacao_CDU.data_emprestimo.dt.year
  zila_mamede_emprestimo_aluno_graduacao_CDU['mes'] = zila_mamede_emprestimo_aluno_graduacao_CDU.data_emprestimo.dt.month
  zila_mamede_emprestimo_aluno_graduacao_CDU = zila_mamede_emprestimo_aluno_graduacao_CDU.drop('data_emprestimo',axis=1)
  zila_mamede_emprestimo_aluno_graduacao_CDU = zila_mamede_emprestimo_aluno_graduacao_CDU.value_counts().reset_index()
  zila_mamede_emprestimo_aluno_graduacao_CDU.columns = ['ano','mes','quantidade']
  zila_mamede_emprestimo_aluno_graduacao_CDU = zila_mamede_emprestimo_aluno_graduacao_CDU.drop('mes',axis=1)
  
  return zila_mamede_emprestimo_aluno_graduacao_CDU
```
